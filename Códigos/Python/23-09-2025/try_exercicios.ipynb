Observe os códigos abaixo, descubra seus erros e trate-os utilizando o bloco ```try-except```. Não é 
para mudar o código fonte, apenas impedir que aconteça um erro fatal no programa.


````python
a = 5
try:
    c = a + b
except NameError:
    print("Variável não declarada")
else:
    print(c)
````

````python
a = 5
b = '5'
try:
    c = a + b
except TypeError:
    print("Tipos incompatíveis para operação")
else:
    print(c)
````

````python
a = 5
b = 5
c = a + b
try:
    print('Resultado: {}'.format())
except IndexError:
    print("Erro ao exibir a variável")
````

````python
a = 5
b = 5
c = a + b
try:
    print("Soma de {} + {} : {}".format(0, 1))
except IndexError:
    print("Erro ao exibir as variáveis")
````

````python
lista = []

try:
    print(lista[0])
except IndexError:
    print("Erro na indexação da lista")
````

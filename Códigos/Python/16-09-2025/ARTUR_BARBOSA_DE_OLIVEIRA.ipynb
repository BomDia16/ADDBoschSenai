### Tarefa 1: Sistema de Priorização de Manutenção Preditiva (```match-case```)

_Cenário_: A Bosch utiliza um sistema automatizado para priorizar chamados de manutenção com base no código de severidade reportado pelos sensores.

_Instruções_:
Crie um script Python que receba um ```codigo_severidade``` (string) como entrada (usar recurso para deixar tudo em caixa-alta se o usuário digitar em caixa baixa) e determine a ação de prioridade. Use a instrução ```match-case``` para implementar a seguinte lógica:

* Se ```codigo_severidade``` for ```"CRITICO"```, imprima: "Prioridade MÁXIMA: Intervenção imediata necessária para evitar falha total."

* Se ```codigo_severidade``` for ```"ALERTA_ALTO"```, imprima: "Prioridade ALTA: Agendar manutenção em até 24 horas. Risco de impacto na produção."

* Se ```codigo_severidade``` for ```"AVISO"```, imprima: "Prioridade MÉDIA: Agendar manutenção preventiva para a próxima janela disponível. Risco baixo."

Para qualquer outro ```codigo_severidade```, imprima: "Status: Normal. Nenhuma ação de manutenção prioritária necessária."

Entrada: "CRITICO"
Saída: Prioridade MÁXIMA: Intervenção imediata necessária para evitar falha total.

Entrada: "INFO"
Saída: Status: Normal. Nenhuma ação de manutenção prioritária necessária.

```````python
# Inserir o código
codigo_severidade = str(input("Insira o código de severidade:"))
codigo_severidade = codigo_severidade.upper()

# Tentar achar o valor inserido e mostrar uma mensagem
match codigo_severidade:
    case 'CRITICO':
        print("Prioridade MÁXIMA: Intervenção imediata necessária para evitar falha total.")
    case 'ALERTA_ALTO':
        print("Prioridade ALTA: Agendar manutenção em até 24 horas. Risco de impacto na produção.")
    case 'AVISO':
        print("Prioridade MÉDIA: Agendar manutenção preventiva para a próxima janela disponível. Risco baixo.")
    case _:
        print("Status: Normal. Nenhuma ação de manutenção prioritária necessária.")
```````

###  Tarefa 2: Classificação de Qualidade de Componentes Eletrônicos (```match-case``` com Múltiplos Padrões)

_Cenário_: Componentes eletrônicos recebidos de fornecedores são classificados por um sistema de controle de qualidade baseado em um código numérico.

_Instruções_:
Desenvolva um script Python que recebe um ``````codigo_qualidade`````` (inteiro) e imprime uma mensagem descrevendo o status do lote. Utilize ```match-case``` com múltiplos padrões:

* Se ``````codigo_qualidade`````` for 1 ou 2, imprima: "Lote Aprovado: Componentes de alta qualidade. Liberado para produção."

* Se ``````codigo_qualidade`````` for 3, imprima: "Lote em Análise: Necessária inspeção adicional para confirmação."

* Se ```codigo_qualidade``` for 4 ou 5, imprima: "Lote Rejeitado: Componentes com defeitos significativos. Encaminhar para descarte ou reprocessamento."

Para qualquer outro ```codigo_qualidade```, imprima: "Código de Qualidade Inválido. Contatar fornecedor."

Exemplo de Entrada/Saída:

Entrada: 4
Saída: Lote Rejeitado: Componentes com defeitos significativos. Encaminhar para descarte ou reprocessamento.

Entrada: 2
Saída: Lote Aprovado: Componentes de alta qualidade. Liberado para produção.

```````python
# Inserir o código
codigo_qualidade = int(input("Insira o código da qualidade:"))

# Tentar achar o valor inserido e mostrar uma mensagem
match codigo_qualidade:
    case 1 | 2:
        print("Lote Aprovado: Componentes de alta qualidade. Liberado para produção.")
    case 3:
        print("Lote em Análise: Necessária inspeção adicional para confirmação.")
    case 4 | 5:
        print("Lote Rejeitado: Componentes com defeitos significativos. Encaminhar para descarte ou reprocessamento.")
    case _:
        print("Código de Qualidade Inválido. Contatar fornecedor.")
```````

### Tarefa 3: Monitoramento de Consumo de Energia (```while``` com ```if```)

_Cenário_: Uma máquina na fábrica Bosch consome energia. O sistema monitora o consumo em intervalos regulares e emite um alerta se o consumo exceder um limite crítico. A monitorização continua até que o alerta seja emitido ou um número máximo de verificações seja atingido.

_Instruções_:
* Crie um script que simule o monitoramento de consumo de energia.

* Inicialize ``````consumo_atual`````` em ```1500``` Watts e limite_critico em  ```2000``` Watts.

* Inicialize ```verificacoes_realizadas``` em ```0``` e um ```max_verificacoes``` em ```10```.

* Use um loop while que continue enquanto ```verificacoes_realizadas < max_verificacoes``` E ```consumo_atual < limite_critico```.

Dentro do loop:

* Simule um aumento aleatório no ```consumo_atual``` (você pode adicionar um valor fixo ou usar ```random.randint``` da library ```random```). Por exemplo, adicione entre 50 e 150 Watts.

* Imprima o ```consumo_atual``` em cada passo, indicando que é uma verificação.

* Incrementa verificacoes_realizadas.

Após o loop, verifique se o ```consumo_atual``` atingiu ou ultrapassou o ```limite_critico```.

* Se sim, imprima: "ALERTA DE CONSUMO CRÍTICO! Desligar a máquina para verificação."

* Se não (ou seja, o loop terminou por atingir ```max_verificacoes```), imprima: "Consumo de energia dentro dos limites aceitáveis após 10 verificações."

```````python
import random

# Inicializar variáveis
consumo_atual = 1500
limite_critico = 2000
verificacoes_realizadas = 0
max_verificacoes = 10

# Verificar os valores
while verificacoes_realizadas < max_verificacoes and consumo_atual < limite_critico:
    consumo_atual = consumo_atual + random.randint(20, 100)
    verificacoes_realizadas += 1
    print(f"Verificação {verificacoes_realizadas}: Consumo atual: {consumo_atual}")

# Verificar se o consumo ultrapassou o limite crítico 
if consumo_atual > limite_critico:
    print("ALERTA DE CONSUMO CRÍTICO! Desligar a máquina para verificação.")
else:
    print("Consumo de energia dentro dos limites aceitáveis após 10 verificações.")
```````

### Tarefa 4: Processamento em Lotes de Peças Usando ```for``` e Verificação de Qualidade

_Cenário_: Um lote de 10 peças foi produzido. Cada peça tem um código de lote. É preciso verificar se alguma peça contém um código de lote "defeituoso".

_Instruções_:
Crie um script Python.

* Crie uma lista ```codigos_pecas``` contendo 10 códigos de lote. Inclua o código ```"ERRO_LOTE_XYZ" ```em alguma posição da lista (por exemplo, na 7ª posição).

* Use um loop for para iterar sobre a lista ```codigos_pecas```.

* Dentro do loop, use uma condicional ```if``` para verificar se o ```codigo_da_peca``` atual é igual a ```"ERRO_LOTE_XYZ"```.

* Se for encontrado o código de erro, imprima: "ERRO DETETADO: Peça com código 'ERRO_LOTE_XYZ' encontrada. Processo de inspeção interrompido." e utilize ```break``` para sair do loop.

* Caso contrário, imprima: "Verificando peça: ```codigo_da_peca```. OK."

Após o loop, use uma flag (variável booleana) para indicar se o erro foi encontrado. Se o loop terminar sem encontrar o erro, imprima: "Lote de peças verificado. Nenhum erro de lote crítico detectado."

```````python
# Inicializar as variáveis
codigos_pecas = ["ERRO_1", "ERRO_2", "ERRO_3", "ERRO_4", "ERRO_5", "ERRO_6", "ERRO_LOTE_XYZ", "ERRO_8", "ERRO_9", "ERRO_10"]
erro_detectado = False

# Verificar se existe o código ERRO_LOTE_XYZ
for codigo in codigos_pecas:
    if codigo == "ERRO_LOTE_XYZ":
        print("ERRO DETETADO: Peça com código 'ERRO_LOTE_XYZ' encontrada. Processo de inspeção interrompido.")
        erro_detectado = True
        break
    else:
        print(f"Verificando peça: {codigo}. OK")
 
# Verificar se não existe nada de errado       
if erro_detectado == False:
    print("Lote de peças verificado. Nenhum erro de lote crítico detectado.")
```````

### Tarefa 5: Gestão de Inventário de Ferramentas (```for``` com ```if``` e Listas)

_Cenário_: A oficina da Bosch mantém um inventário de ferramentas. Precisamos verificar se todas as ferramentas essenciais estão presentes antes do início de um turno.

_Instruções_:

Crie uma lista chamada ```ferramentas_essenciais``` com os seguintes itens: ```['Chave de Fenda', 'Alicate Universal', 'Chave Inglesa', 'Martelo', 'Multímetro']```.

Crie uma lista chamada ```ferramentas_disponiveis``` que contenha algumas das ferramentas essenciais e talvez uma ou duas extras (ex: ```['Chave de Fenda', 'Chave Inglesa', 'Alicate Bomba D\'água', 'Martelo']```).

Crie uma flag ```todas_presentes``` inicializada como ```True```.

Use um loop ```for``` para iterar sobre a lista ```ferramentas_essenciais```.

Dentro do loop, use uma condicional ```if``` para verificar se a ```ferramenta_essencial``` atual não está na lista ```ferramentas_disponiveis```.

Se a ferramenta não estiver presente, imprima: "FALTA FERRAMENTA: ```ferramenta_essencial``` não encontrada no inventário."

Defina ```todas_presentes``` como ```False```.

Após o loop, verifique o valor da flag ```todas_presentes```. Se for ```True```, imprima: "Inventário Completo: Todas as ferramentas essenciais estão disponíveis."

```````python
# Inicializar as variáveis
ferramentas_essenciais = ['Chave de Fenda', 'Alicate Universal', 'Chave Inglesa', 'Martelo', 'Multímetro']
ferramentas_disponiveis = ['Chave de Fenda', 'Chave Inglesa', 'Alicate Bomba D\'água', 'Martelo']
todas_presentes = True

# Verificar se todas as ferramentas essenciais estão disponíveis
for essencial in ferramentas_essenciais:
    if essencial not in ferramentas_disponiveis:
        print(f"FALTA FERRAMENTA: {essencial} não encontrada no inventário.")
        todas_presentes = False

# Caso todas estejam presentes, mostrar a mensagem       
if todas_presentes:
    print("Inventário Completo: Todas as ferramentas essenciais estão disponíveis.")
```````

### Tarefa 6: Processamento de Ordens de Serviço de Equipamentos (```while``` e ```for```)

_Cenário_: Várias ordens de serviço de manutenção para diferentes equipamentos da Bosch chegam ao sistema. Cada ordem tem um ID e uma lista de tarefas. Precisamos processar essas ordens sequencialmente.

_Instruções:_

1. Crie uma lista de dicionários, onde cada dicionário representa uma ```ordem_servico```. Cada ordem deve ter:

* ```id_ordem``` (ex: 1001, 1002)
* equipamento (ex: "Linha de Montagem A", "Máquina de Solda B")

* tarefas (uma lista de strings, ex: ```['Verificar rolamentos', 'Lubrificar engrenagens']```)

2. Crie uma variável ```ordens_a_processar``` inicializada com o número de ordens que você criou (por exemplo, 2).

3. Use um loop ```while``` que continue enquanto ```ordens_a_processar > 0```.

4. Dentro do ```while```:

* Obtenha a próxima ordem de serviço da sua lista (você pode usar o índice ou ```pop()``` para remover e processar).

* Imprima o ID da ordem e o nome do equipamento.

* Use um loop ```for``` para iterar sobre a lista de tarefas da ordem atual e imprima cada tarefa.

* Após processar todas as tarefas de uma ordem, imprima uma mensagem de ```"Ordem de Serviço #id_ordem concluída."```

* Decremente ```ordens_a_processar```.

5. Quando o loop ```while``` terminar, imprima: ```"Todas as ordens de serviço foram processadas."```

```````python
# Inicializar as variáveis
ordem_servico = [
    {
        'id_ordem': 1001,
        'equipamento': 'Linha de Montagem A',
        'tarefas': ['Verificar rolamentos', 'Lubrificar engrenagens']
    },
    {
        'id_ordem': 1002,
        'equipamento': 'Linha de Montagem B',
        'tarefas': ['Verificar pastilhas', 'Lubrificar peças']
    }
]

ordens_a_processar = len(ordem_servico)

# Processar cada uma das ordens
while ordens_a_processar > 0:
    for ordem in ordem_servico:
        print(f"Processando Ordem de Serviço ID: {ordem['id_ordem']} - Equipamento: {ordem['equipamento']}")
        for tarefa in ordem['tarefas']:
            print(f"    - Tarefa: {tarefa}")
            
        print(f"Ordem de Serviço #{ordem['id_ordem']} concluída.")
        ordens_a_processar -= 1
        
print("Todas as ordens de serviço foram processadas.")
```````

### Tarefa 7: Relatório de Vendas por Departamento (```for``` e Dicionários)

_Cenário:_ A área de vendas da Bosch agrupa os relatórios de vendas por departamento. Precisamos consolidar essas informações.

_Instruções:_

Crie um dicionário chamado ```vendas_departamentos```. As chaves deste dicionário serão os nomes dos departamentos (ex: "Automotivo", "Industrial", "Eletrônicos") e os valores serão listas de inteiros representando vendas diárias em uma semana.

* Exemplo: ```{'Automotivo': [1200, 1500, 1350, 1600, 1400, 1700, 1550], 'Industrial': [800, 950, 880, 1000, 920, 1100, 1050]}```

* Use um loop for para iterar sobre as chaves (departamentos) do dicionário ```vendas_departamentos```.

* Dentro do loop, para cada departamento:

* Calcule a soma total das vendas da semana na lista de vendas correspondente.

* Imprima o nome do departamento e o total de vendas da semana formatado.

```````python
# Inicializar as variáveis
vendas_departamentos = {
    'Automotivo': [1200, 1500, 1350, 1600, 1400, 1700, 1550],
    'Industrial': [800, 950, 880, 1000, 920, 1100, 1050],
    'Eletrônicos': [900, 150, 980, 2000, 120, 2100, 2050],
}

# Somar as vendas de cada departamento
for departamento in vendas_departamentos:
    soma = sum(vendas_departamentos[departamento])
    print(f"Departamento: {departamento} - Total de Vendas da Semana: R$ {soma:,.2f}")
```````

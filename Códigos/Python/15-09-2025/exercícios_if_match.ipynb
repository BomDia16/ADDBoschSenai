# Tarefa 1: Controle de Qualidade de Peças (Condicional Simples)
Cenário: A Bosch utiliza sensores para medir a dimensão de peças produzidas. Uma peça é
considerada aprovada se sua dimensão estiver dentro de um intervalo de tolerância. Caso contrário,
ela é rejeitada.

1. Crie uma variável dimensao_peca e atribua a ela um valor numérico (por exemplo, 15.2).
2. Defina o valor mínimo de tolerância como 15.0 e o máximo como 15.5.
3. Use uma condicional simples para verificar se a dimensao_peca está entre o mínimo e o
máximo (inclusive).
4. Se a condição for verdadeira, imprima a mensagem: "Peça Aprovada: dentro da tolerância."
5. Se a condição for falsa, imprima: "Peça Rejeitada: fora da tolerância."

```python
dimensao_peca = 15.2

tolerancia_minima = 15
tolerancia_maxima = 15.5

if dimensao_peca > tolerancia_minima and dimensao_peca <= tolerancia_maxima:
    print("Peça Aprovada: dentro da tolerância.")
else:
    print("Peça Rejeitada: fora da tolerância.")
```

# Tarefa 2: Análise de Produtividade do Robô (Condicional Aninhado)
Cenário: Um robô da linha de montagem da Bosch tem sua eficiência avaliada com base em dois
critérios: a quantidade de peças processadas por hora e o percentual de peças defeituosas. Para ser
considerado altamente produtivo, o robô deve processar mais de 100 peças por hora e ter um
percentual de defeitos inferior a 2%. Se ele processa mais de 100 peças, mas o percentual de
defeitos é maior ou igual a 2%, sua produtividade é considerada "Média". Se ele processa 100 ou
menos peças, sua produtividade é "Baixa".
Instruções:
1. Crie as variáveis pecas_por_hora (por exemplo, 120) e percentual_defeitos (por exemplo,
1.5).
2. Use uma condicional aninhada para avaliar a produtividade do robô.
3. Primeiro, verifique se pecas_por_hora é maior que 100.
    o Se for, use uma condicional aninhada para verificar se percentual_defeitos é menor
que 2.
    ▪ Se sim, imprima: "Produtividade do Robô: Alta."
    ▪ Se não, imprima: "Produtividade do Robô: Média."
4. Se a primeira condição (peças > 100) for falsa, imprima: "Produtividade do Robô: Baixa."

```python
pecas_por_hora = 120
percentual_defeitos = 1.5

if pecas_por_hora > 100:
    if percentual_defeitos < 2:
        print("Produtividade do Robô: Alta.")
    else:
        print("Produtividade do Robô: Média.")

else:
    print("Produtividade do Robô: Baixa.")
```

# Tarefa 3: Sistema de Alerta de Manutenção Preditiva (Condicional Simples e Aninhado)
Cenário: A Bosch utiliza sensores IoT para monitorar a temperatura e a vibração de uma máquina
industrial. O sistema de manutenção preditiva deve emitir um alerta com base nas seguintes regras:
• Alerta Máximo: Se a temperatura for superior a 90°C e a vibração for superior a 7.5.
• Alerta Médio: Se a temperatura for superior a 90°C, mas a vibração for inferior ou igual a
7.5.
• Alerta de Vibração: Se a temperatura for inferior ou igual a 90°C e a vibração for superior a
7.5.
• Status OK: Se a temperatura for inferior ou igual a 90°C e a vibração for inferior ou igual a
7.5.
Instruções:
1. Crie as variáveis temperatura_motor (ex: 95.5) e vibracao_motor (ex: 8.1).
2. Construa um conjunto de condicionais (if, elif, else) para cobrir todas as regras acima.
3. Use a combinação de if com operadores lógicos (and, or) para testar as condições.

```python
temperatura_motor = 90
vibracao_motor = 8.1

if temperatura_motor > 90 and vibracao_motor > 7.5:
    print("Alerta Máximo")
elif temperatura_motor > 90 and vibracao_motor <= 7.5:
    print("Alerta Médio")
elif temperatura_motor <= 90 and vibracao_motor > 7.5:
    print("Alerta de Vibração")
elif temperatura_motor <= 90 and vibracao_motor <= 7.5:
    print("Status OK")
```

# Tarefa 4: Classificação de Falhas de Componentes (Match Case)
Cenário: Na linha de produção da Bosch, sensores em componentes eletrônicos podem gerar
códigos de erro específicos. Um sistema de diagnóstico precisa classificar esses códigos para
direcionar a equipe de manutenção.
Instruções:
1. Crie uma variável codigo_erro e atribua a ela um dos seguintes valores: '
2. E01', 'E02', 'E03', 'W01', ou outro valor qualquer.
3. Utilize a estrutura match case para classificar o erro:
o Se codigo_erro for 'E01', imprima: "ERRO CRÍTICO: Falha de comunicação principal.
Parar linha de produção imediatamente."
o Se codigo_erro for 'E02', imprima: "ERRO GRAVE: Superaquecimento detectado.
Desligar componente."
o Se codigo_erro for 'E03', imprima: "ERRO PADRÃO: Diodo emissor de luz falhou.
Substituir componente."
o Se codigo_erro for 'W01', imprima: "ALERTA: Baixo nível de energia. Verificar fonte
de alimentação."
o Para qualquer outro valor de codigo_erro, utilize o case _ (wildcard) para imprimir:
"Status: Operação normal ou código de erro desconhecido."

```python
codigo_erro = 'E01'
match codigo_erro:
    case 'E01':
        print("ERRO CRÍTICO: Falha de comunicação principal. Parar linha de produção imediatamente.")
    case 'E02':
        print("ERRO GRAVE: Superaquecimento detectado. Desligar componente.")
    case 'E03':
        print("ERRO PADRÃO: Diodo emissor de luz falhou. Substituir componente.")
    case 'W01':
        print("ALERTA: Baixo nível de energia. Verificar fonte de alimentação.")
    case _:
        print("Status: Operação normal ou código de erro desconhecido.")
```

# Tarefa 5: Níveis de Criticidade de Manutenção (Match Case com Agrupamento)
Cenário: Em uma planta da Bosch, as ordens de serviço de manutenção são priorizadas com base na
criticidade do equipamento. A criticidade é representada por um número.
Instruções:
1. Crie uma variável nivel_criticidade e atribua a ela um valor numérico inteiro (por exemplo,
3).
2. Utilize a estrutura match case para determinar a ação apropriada:
o Se nivel_criticidade for 1 ou 2, imprima: "Prioridade MÁXIMA: Manutenção imediata
requerida."
o Se nivel_criticidade for 3 ou 4, imprima: "Prioridade ALTA: Agendar manutenção nas
próximas 24 horas."
o Se nivel_criticidade for 5, imprima: "Prioridade MÉDIA: Agendar manutenção na
próxima semana."
o Para qualquer outro valor numérico ou não numérico, utilize o case _ para imprimir:
"Status: Manutenção de rotina ou nível inválido."

```python
nivel_criticidade = 3

match nivel_criticidade:
    case 1 | 2:
        print("Prioridade MÁXIMA: Manutenção imediata requerida.")
    case 3 | 4:
        print("Prioridade ALTA: Agendar manutenção nas próximas 24 horas.")
    case 5:
        print("Prioridade MÉDIA: Agendar manutenção na próxima semana.")
    case _:
        print("Status: Manutenção de rotina ou nível inválido.")
```

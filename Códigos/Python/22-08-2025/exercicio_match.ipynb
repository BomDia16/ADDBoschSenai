1. Crie um programa para responder a questão, usando strings:

Qual linguagem de programação que tem domínio? 
* python: Linguagem muito utilizada em análise de dados e IoT
* sql: Linguagem utilizada em Banco de Dados
* javascript: Muito usada em Desenvolvimento web
* nenhuma das anteriores: Não importa a linguagem, importa resolver problemas
      

```python

print("python")
print("sql")
print("javascript")
print("nenhuma das anteriores")
linguagem = str(input("Qual linguagem de programação que tem domínio?"))
linguagem = linguagem.lower()

match linguagem:
    case "python":
        print("Linguagem muito utilizada em análise de dados e IoT")
    case "sql":
        print("Linguagem utilizada em Banco de Dados")
    case "javascript":
        print("Muito usada em Desenvolvimento web")
    case "nenhuma das anteriores":
        print("Não importa a linguagem, importa resolver problemas")
    case _:
        print("Não importa a linguagem, importa resolver problemas")
        
```

2.Desenvolva um "Tradutor de Emoções" que peça ao usuário para inserir uma emoção (feliz, triste, nervoso etc.). Utilize a estrutura match..case para imprimir um emoji correspondente a essa emoção. Por exemplo, imprimir :) para feliz, :( para triste etc.

```python
print("Feliz")
print("Triste")
print("Nervoso")
print("Irritado")
emocao = str(input("Qual é a sua emoção hoje?"))
emocao = emocao.lower()

match emocao:
    case "feliz":
        print("=)")
    case "triste":
        print("=(")
    case "nervoso":
        print("<=(")
    case "irritado":
        print(">=(")
    case _:
        print("Não tenho essa emoção no meu banco de dados")
```

3. Escreva um programa que peça ao usuário para inserir uma faixa etária e, em seguida, utilize a estrutura match..case para sugerir um filme adequado para essa idade. Você pode categorizar as faixas etárias e sugerir diferentes tipos de filmes para cada categoria, como filmes infantis para crianças, ação para adolescentes, drama para adultos etc.

```python
idade = int(input("Insira a sua idade:"))
faixa_etaria = ""

if (idade <= 10):
    faixa_etaria = "infantil"
elif (idade < 18):
    faixa_etaria = "adolescente"
else:
    faixa_etaria = "adulto"
    
match faixa_etaria:
    case "infantil":
        print("Assista Toy Story")
    case "adolescente":
        print("Assista Gente Grande")
    case "adulto":
        print("Assista IT a coisa")
```

4. Crie um programa que peça ao usuário para inserir dois números, a e b, e uma operação aritmética (+, -, * ou /). Use a estrutura match..case para avaliar uma expressão que combine a operação inserida pelo usuário e, em seguida, imprima o resultado da operação escolhida.

```python
n1 = int(input("Insira um número inteiro:"))
n2 = int(input("Insira um outro número inteiro:"))

print("Soma")
print("Subtração")
print("Multiplicação")
print("Divisão")
operacao = str(input("Qual operação gostaria de realizar?"))

operacao = operacao.lower()

resultado = 0

match operacao:
    case "soma":
        resultado = n1 + n2
    case "subtração":
        resultado = n1 - n2
    case "multiplicação":
        resultado = n1 * n2
    case "divisão":
        if n2 == 0:
            print("Não é possível realizar divisão por zero")
        else:
            resultado = n1 / n2

print("O resultado da conta é", resultado)
```

6. Crie um programa que peça a idade do usuário e use  para determinar a sua categoria.
Categorias:
* 0-12: Criança
* 13-17: Adolescente
* 18-59: Adulto
* 60+: Idoso
* Qualquer outro valor: Idade inválida

Dica: Usar if ternário em cada case

```python
idade = int(input("Insira a sua idade:"))

categoria = (
    "Criança" if idade >= 0 and idade <= 12 else
    "Adolescente" if idade >= 13 and idade <= 17 else
    "Adulto" if idade >= 18 and idade <= 59 else
    "Idoso" if idade >= 60 else "Idade Inválida"
)

print(categoria)
```

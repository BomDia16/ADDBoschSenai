Notas

```python
nota = 7

if (nota <= 4):
    print("Reprovado")
elif (nota > 4 and nota < 7):
    print("Recuperação")
else:
    print("Aprovado")
```

Ver se o numero está entre 100 e 200

```python
num = int(input("Digite um número:"))

if (num > 100 and num < 200):
    print(f"O número {num} está entre 100 e 200")
else:
    print(f"O número {num} não está entre 100 e 200")
```

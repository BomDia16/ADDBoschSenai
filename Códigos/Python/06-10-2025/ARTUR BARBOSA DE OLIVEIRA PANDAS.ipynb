## Usar a biblioteca pandas para as tarefas seguintes. Usar o arquivo linha_producao.txt

Determinar a dimensão do DataFrame (linhas, colunas).

```python
# importar o pandas
import pandas as p

# abrir o arquivo de dados
mydataset = p.read_csv("linha_producao.txt", sep=",")

# valor de linhas e colunas da tabela
print(mydataset.shape)
```

Obter um resumo completo da estrutura, tipos de dados (dtypes) e contagem de valores não-nulos. Fundamental para identificar colunas com valores ausentes.

```python
# informações da tabela
print(mydataset.info())
```

Visualizar as primeiras 5 linhas para uma inspeção rápida.

```python
# 5 primeiras linhas
print(mydataset.head())
```

Visualizar as últimas 3 linhas (exemplo com 3).

```python
# 3 últimas linhas
print(mydataset.tail(3))
```

Gerar estatísticas descritivas (média, desvio padrão, mínimo, quartis).

```python
# descrição da tabela
print(mydataset.describe())
```

Remover todas as linhas que contenham pelo menos um valor nulo (NaN). Cria um novo DataFrame 'df_limpo'.

```python
# limpa as linhas que possuam valores nulos
df_limpo = mydataset.dropna()
print(df_limpo)
```

Intervalo de datas no índice (Timestamp) e todas as colunas. usar loc() para as próximas tarefas

```python
# trazer todos os valores da coluna Timestamp, ou seja, todas as datas
tempos = mydataset["Timestamp"]
print(tempos)
```

Valor da coluna Peças_Produzidas para um único Timestamp.

```python
# trará os valores da linha no dia 2 de janeiro de 2024 as 7 da manhã
jan_2_2024 = mydataset.loc[mydataset["Timestamp"] == "2024-01-02 07:00:00"]

# mostrará na tela apenas o valor de peças no dia resgatado
print(jan_2_2024["Peças_Produzidas"])
```

Todas as linhas da Linha_Producao igual a 'Linha_A' com 'Status_Qualidade' de 'Defeito_A'. Usar & para and.

```python
# usa o operador & para encontrar linhas que são da linha A de produção e possuam o defeito A
linha_A = mydataset.loc[(mydataset["Linha_Producao"] == "Linha_A") & (mydataset["Status_Qualidade"] == "Defeito_A")]
print(linha_A)
```

Selecionar apenas as colunas de temperatura e vibração para a Linha B.

```python
# encontra todas as linhas da tabela que sejam da Linha B de produção
linha_B = mydataset.loc[mydataset["Linha_Producao"] == "Linha_B"]

# selecionar apenas as colunas de temperatura e de vibração
print(linha_B[["Sensor_Temperatura_Máquina", "Sensor_Vibração_Máquina"]])
```

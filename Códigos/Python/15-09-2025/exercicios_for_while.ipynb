# Tarefa 1: Monitoramento de Nível de Óleo (Loop while com if)
Cenário: Um sensor na linha de produção da Bosch monitora continuamente o nível de óleo em
uma máquina. O nível ideal está entre 50% e 75%. Se o nível cair abaixo de 50%, um alerta é emitido
e a máquina deve parar. Se o nível estiver acima de 75%, outro tipo de alerta é emitido, mas a
máquina continua operando. O monitoramento deve parar apenas quando o nível atingir um valor
crítico ou a máquina for parada manualmente.
Instruções:
1. Inicialize uma variável nivel_oleo com um valor inicial (ex: 60.0).
2. Defina os limites de alerta: limite_inferior = 50.0 e limite_superior = 75.0.
3. Inicie com: Iniciando monitoramento do nível de óleo...
4. Crie um loop while que continue executando enquanto o nivel_oleo for maior que 0
(simulando o monitoramento contínuo).
5. Dentro do loop, use um if para verificar o nivel_oleo:
o Se nivel_oleo < limite_inferior:
▪ Imprima "ALERTA CRÍTICO: Nível de óleo baixo {nível_oleo}! Máquina será
parada."
▪ Use break para sair do loop.
o Se nivel_oleo > limite_superior:
▪ Imprima “ATENÇÃO: Nível de óleo alto {nivel_oleo). Monitorando."
o Se nenhuma das condições acima for verdadeira (ou seja, o nível está entre 50% e
75%), imprima "Nível de óleo OK: {nivel_oleo}."

6. Após cada verificação de nível, simule uma pequena variação no nivel_oleo (ex: subtraia um
valor pequeno, como 1.5) para que o loop avance.

```python
nivel_oleo = 60
limite_inferior = 50
limite_superior = 75

print("Iniciando monitoramento do nível de óleo...")
while nivel_oleo > 0:
    if nivel_oleo < limite_inferior:
        print(f"ALERTA CRÍTICO: Nível de óleo baixo {nivel_oleo}! Máquina será parada.")
        break
    elif nivel_oleo > limite_superior:
        print(f"ATENÇÃO: Nível de óleo alto {nivel_oleo}. Monitorando.")
        break
    else:
        print(f"Nível de óleo OK: {nivel_oleo}.")
    
    nivel_oleo = nivel_oleo - 1.5
```

# Tarefa 2. Inspeção de Componentes em Lote (Loop for com if)
Cenário: A Bosch recebe lotes de componentes eletrônicos para montagem. Cada componente é
inspecionado. O lote é considerado aprovado se a maioria dos componentes estiver dentro da
tolerância, mas qualquer componente com defeito grave deve ser imediatamente descartado e o
processo interrompido.
Instruções:
1. Crie uma lista chamada lote_componentes contendo valores numéricos que representam a
qualidade de cada componente ([9.8, 9.5, 9.9, 8.5, 9.7, 9.0, 10.0, 7.0, 9.6]).
2. Defina um valor de tolerância mínima para aprovação (tolerancia_minima = 8.0) e um limite
para defeito grave (defeito_grave = 7.5).
3. Inicialize um contador componentes_aprovados = 0.
4. Use um loop for para iterar sobre cada componente na lista lote_componentes.
5. Dentro do loop, use um if para verificar cada componente:
o Se componente < defeito_grave:
▪ Imprima “Componente com defeito grave encontrado: {componente}.
Interrompendo inspeção do lote."
▪ Use break para sair do loop.
o Se componente >= tolerancia_minima:
▪ Imprima "Componente {componente} dentro da tolerância. Contando como
aprovado."
▪ Incremente.
o Caso contrário (se o componente não for gravemente defeituoso nem totalmente
aprovado), imprima "Componente {componente} fora da tolerância ideal, mas
aceitável."

6. Após o loop, verifique se o loop foi interrompido por break ou se todos os componentes
foram processados. Uma forma de fazer isso é usando um else associado ao for (que executa
apenas se o loop terminar sem break). Dentro desse else, você pode imprimir um resumo
ou a decisão final sobre o lote.
o Se o else do for for executado, imprima: "Inspeção do lote concluída. Total de
componentes aprovados: {componentes_aprovados} de {lote_componentes}."

```python
lote_componentes = [9.8, 9.5, 9.9, 8.5, 9.7, 9.0, 10.0, 7.0, 9.6]
tolerancia_minima = 8.0
defeito_grave = 7.5
componentes_aprovados = 0

for componente in lote_componentes:
    if componente < defeito_grave:
        print(f"Componente com defeito grave encontrado: {componente}. Interrompendo inspeção do lote.")
        break
    elif componente >= tolerancia_minima:
        print(f"Componente {componente} dentro da tolerância. Contando como aprovado.")
        componentes_aprovados += 1
    else:
        print(f"Componente {componente} fora da tolerância ideal, mas aceitável.")
        
else:
    print("\nInspeção do lote concluída sem defeitos graves.")
    print(f"Total de componentes aprovados: {componentes_aprovados} de {len(lote_componentes)}.")

print("Inspeção de lote finalizada.")
```

# Questão 5: Análise de Código de Produtos
Você tem uma lista de códigos de produtos, mas alguns estão desatualizados e precisam ser
substituídos.

1. Crie uma lista de códigos de produtos codigos_antigos = ['01B', '02C', '03A', '04D', '05E'].

```python
codigos_antigos = ['01B', '02C', '03A', '04D', '05E']
```

2. Crie uma lista de novos códigos novos_codigos = ['06F', '07G'].

```python
novos_codigos = ['06F', '07G']
```

3. Use extend() para adicionar os novos códigos à lista de códigos antigos.

```python
codigos_antigos.extend(novos_codigos)
print(codigos_antigos)
```

4. O código '03A' deve ser substituído por um novo código 03H. Use remove() para tirar o '03A'
e, em seguida, use insert() para colocar o '03H' na mesma posição que ele ocupava (índice
2).

```python
codigos_antigos.remove('03A')
codigos_antigos.insert(2, '03H')
print(codigos_antigos)
```

5. O código '04D' foi retirado do catálogo. Use pop() para removê-lo da lista e armazene o valor
removido em uma variável.

```python
removido = codigos_antigos.pop(3)
print(codigos_antigos)
```

6. Conte quantas vezes o caractere 'B' aparece nos códigos da lista final. (Dica: você pode
precisar de um loop para iterar sobre os itens da lista e usar o método count() em cada
string).

```python
contador = 0

for i in codigos_antigos:
    contador += i.count('B')
    
print(f"Tem {contador} 'B' nos códigos")
```

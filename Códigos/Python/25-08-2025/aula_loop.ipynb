# While

Contar de 1 até 10

```python
contador = 1
while contador <= 10:
    print(contador)
    contador += 1
```

Repetir SENAI 3 vezes junto com a contagem

```python
contador = 1
while contador <= 3:
    print(f"{contador}. SENAI")
    contador += 1
```

Pedir os três primeiros nomes e sobrenomes, separados, use concatenação para imprimir na tela

```python
contador = 1
while contador <= 3:
    nome = str(input("Insira o nome:"))
    sobrenome = str(input("Insira o sobrenome:"))
    nome = nome.capitalize()
    sobrenome = sobrenome.capitalize()
    
    nome_completo = nome + " " + sobrenome
    print(nome_completo)
    contador += 1
    
```

Menu para saber dia da semana

```python
while True:
    opcao = int(input("Insira um número inteiro:"))
    
    match opcao:
        case 1:
            print("Domingo")
        case 2:
            print("Segunda-Feira")
        case 3:
            print("Terça-Feira")
        case 4:
            print("Quarta-Feira")
        case 5:
            print("Quinta-Feira")
        case 6:
            print("Sexta-Feira")
        case 7:
            print("Sábado")
        case _:
            print("Até mais")
            break
```

# For

Contar até 10

```python
for i in range(1, 11):
    print(i)
```

Contar de 10 a 1

```python
for i in range(10, 0, -1):
    print(i)
```

# 2 Cálculo da Área do Círculo
• Enunciado: Desenvolva uma função que receba o raio de um círculo e retorne sua área
(A=πr2). (Deve-se utilizar a constante π do módulo math).

```python
import math as m

def area_circulo(raio):
    try:
        area = m.pow(raio, 2) * m.pi
    except TypeError:
        print("Digite um número para o raio")
    else:
        return area 

print(f"O círculo com raio 5 tem área de {area_circulo(5)}")
print(f"{area_circulo("f")}")
```

# 3 Verificação de Número Par ou Ímpar
• Enunciado: Crie uma função que receba um número inteiro e verifique se ele é par ou
ímpar, imprimindo o resultado dessa verificação.

```python
def par_impar(numero):
    try:
        if numero % 2 == 0:
            return "par"
        else:
            return "ímpar"
    except TypeError:
        print("Digite um número para a verificação")
        
print(par_impar(4))
print(par_impar(5))
print(par_impar("a"))
```

# 4 Inversão de uma String
• Enunciado: Escreva uma função que receba uma string (texto) e retorne essa mesma string,
mas com os caracteres em ordem inversa.

```python
def inversor(texto):
    try:
        texto = "".join(reversed(texto))
        print(texto)
    except TypeError:
        print("Digite um número para reverter")

inversor(123)
```

# 5 Cálculo da Média de uma Lista de Números
• Enunciado: Implemente uma função que receba uma lista de números e calcule e retorne a média aritmética desses números.

```python
def media_lista(lista):
    try:
        media = sum(lista) / len(lista)
        print(f"{media:.2f}")
    except TypeError:
        print("Digite um array de números para conseguir a média")

media_lista([2, 5, 10])
media_lista("nkjfg")
```

# 6 Gerar Sequência de Fibonacci
• Enunciado: Crie uma função que receba um limite (um número máximo) e retorne uma lista contendo a sequência de Fibonacci até aquele limite.

```python
def fibonacci_iterativo(n):
    try:
        sequencia = []
        a, b = 0, 1
        while len(sequencia) < n:
            sequencia.append(a)
            a, b = b, a + b
        return sequencia
    except TypeError:
        print("Digite um número para receber a quantidade elementos da Sequência de Fibonacci")

# Exemplo de uso
num_elementos = 10
print(f"Sequência de Fibonacci com os {num_elementos} primeiros elementos: {fibonacci_iterativo(num_elementos)}")
num_elementos = "gsdg"
print(fibonacci_iterativo(num_elementos))
```

# 7 Contagem de Vogais em um Texto
• Enunciado: Desenvolva uma função que receba uma string e retorne o número total de
vogais (a, e, i, o, u, e suas variações com acento) presentes no texto, ignorando a diferença
entre maiúsculas e minúsculas.

```python
def num_vogais(texto):
    try:
        vogais = "aeiouáâàãéêèêíîìóôòõúûù"
        texto = texto.lower()
        qtd_vogais = 0
        
        for char in vogais:
            for chart in texto:
                if char == chart:
                    qtd_vogais += 1
                
        print(qtd_vogais)
    except Exception:
        print("Insira um texto para verificar a quantidade de vogais")
        
num_vogais("vogaisáAÉ")
num_vogais(9)
```

# 8. Conversão Segura para Percentual
Crie uma função chamada converter_para_percentual que tente receber uma string do usuário
e convertê-la em um número decimal para representar uma porcentagem (onde "50" se torna
0.5).

A função deve usar um bloco try/except para garantir que a entrada seja válida:
1. Tratamento de Erro de Tipo (ValueError): Se a string de entrada não puder ser convertida
em um número (por exemplo, se o usuário digitar "abc" ou deixar a entrada vazia), o bloco
except deve imprimir uma mensagem de erro ("ERRO: Entrada inválida. Digite apenas
números.") e retornar o valor None.
2. Cálculo e Retorno: Se a conversão for bem-sucedida, o número deve ser dividido por 100 e o
resultado (float) deve ser retornado.
Observação: A função deve aceitar apenas um argumento (valor_str).

```python
def converter_para_percentual(percentual):
    try:
        percentual = float(percentual)
    except ValueError:
        print("ERRO: Entrada inválida. Digite apenas números.")
    else:
        decimal = percentual / 100
        print(decimal)
        
converter_para_percentual(65.5)
converter_para_percentual("fdg")
```

# Análise de Lotes de Produção Bosch

Crie uma função em Python chamada analisar_lote_producao que simule a análise de qualidade de
um lote de peças em uma linha de montagem da Bosch.
A função deve aceitar os seguintes argumentos:
1. dados_producao (list of floats): Uma lista contendo as medições de diâmetro (em
milímetros) de uma amostra de peças do lote.
2. especificacao_min (float): O diâmetro mínimo aceitável, conforme as especificações da
engenharia Bosch.
3. especificacao_max (float): O diâmetro máximo aceitável.

Requisitos da Função:
1. Cálculo da Média: Calcule a média das medições de diâmetro no lote.
2. Verificação de Não Conformidade: Conte o número de peças não-conformes (que estão
fora do intervalo [especificacao_min, especificacao_max]).
3. Tratamento de Erros (try/except):
o Utilize um bloco try/except para capturar um possível erro de divisão por zero que
ocorreria se a lista dados_producao estivesse vazia.
o Se a lista estiver vazia, o bloco except deve retornar uma tupla com os valores (0, 0)
e imprimir uma mensagem de erro clara, como: "ERRO: Lista de produção vazia.
Impossível calcular a média."

4. Retorno: A função deve retornar uma tupla contendo dois valores:
o O valor da média dos diâmetros (float).
o O número total de peças não-conformes (int).

```python
def analisar_lote_producao(dados_producao, especificacao_min, especificacao_max):
    try:
        media_dados = sum(dados_producao) / len(dados_producao)
        
        qtd_non_conformities = 0
    
        for medicao in dados_producao:
            if medicao > especificacao_max or medicao < especificacao_min:
                qtd_non_conformities += 1
            
        return media_dados, qtd_non_conformities
    except ZeroDivisionError:
        print("ERRO: Lista de produção vazia. Impossível calcular a média.")
        return 0, 0
    except TypeError:
        print("Insira valores válidos para a análise (números)")
        return 0, 0

print(analisar_lote_producao([13, 15], "10", 20))
print(analisar_lote_producao([13, 15, 21], 10, 20))
print(analisar_lote_producao([13, "15", 21], 10, 20))
```

# Processamento de Dados de Vendas com Validação
Crie uma função em Python chamada processar_registro_vendas que recebe um dicionário
contendo dados de uma única transação de venda e tente calcular o valor total da transação
(quantidade × preço unitário).
O dicionário de entrada pode ter o seguinte formato:
registro = {
"item": "Teclado Mecânico",
"quantidade": "10", # Nota: Recebido como string
"preco_unitario": "150.50" # Nota: Recebido como string
}

Requisitos da Função:
1. Acesso Seguro aos Dados (try/except para Chaves):
o Use um bloco try para tentar acessar as chaves "quantidade" e "preco_unitario" no
dicionário de entrada.
o Se uma dessas chaves não existir no dicionário, capture o erro KeyError.
o No bloco except KeyError, imprima a mensagem: "ERRO DE DADOS: O dicionário de
vendas está incompleto. Faltam as chaves 'quantidade' ou 'preco_unitario'." e
retorne o valor None.

2. Conversão Segura de Valores (try/except para Tipo):
o Após acessar as chaves com sucesso, use um try/except (pode ser o mesmo bloco ou
um aninhado) para tentar converter os valores de quantidade e preco_unitario para
o tipo numérico (float ou int).
o Se a conversão falhar (ex: se quantidade for "dez"), capture o erro ValueError.
o No bloco except ValueError, imprima a mensagem: "ERRO DE VALOR: A quantidade
ou o preço unitário não são números válidos." e retorne o valor None.

3. Cálculo e Retorno:
o Se todos os acessos e conversões forem bem-sucedidos, calcule o valor total da
venda (quantidade × preço unitário).
o A função deve retornar o valor total (float).

```python
def processar_registro_vendas(transacao):
    try:
        quantidade = transacao["quantidade"]
        preco_unitario = transacao["preco_unitario"]
        
        quantidade = int(quantidade)
        preco_unitario = float(preco_unitario)
        
        valor_total = quantidade * preco_unitario
        return valor_total
    except KeyError:
        print("ERRO DE DADOS: O dicionário de vendas está incompleto. Faltam as chaves 'quantidade' ou 'preco_unitario'.")
        return None
    except ValueError:
        print("ERRO DE VALOR: A quantidade ou o preço unitário não são números válidos.")
        return None

registro = {
    "item": "Teclado Mecânico",
    "quantidade": "10", # Nota: Recebido como string
    "preco_unitario": "150.50" # Nota: Recebido como string
}

registro2 = {
    "item": "Teclado Mecânico",
    "quantidade": "GEGE", # Nota: Recebido como string
    "preco_unitario": "150.50" # Nota: Recebido como string
}

registro3 = {
    "item": "Teclado Mecânico",
    "quantidade": "10", # Nota: Recebido como string
    "gdfgdfg": "150.50" # Nota: Recebido como string
}

print(processar_registro_vendas(registro))
print(processar_registro_vendas(registro2))
print(processar_registro_vendas(registro3))
```

If ternário

```python
idade = int(input("Digite a sua idade:"))

status = "Maior" if idade >= 18 else "Menor"

print(status)
```

Digite um inteiro, de 1 a 7, e retorne o dia da semana correspondente

```python
num = int(input("Digite um número inteiro:"))

match num:
    case 1:
        print("Domingo")
    case 2:
        print("Segunda-Feira")
    case 3:
        print("Terça-Feira")
    case 4:
        print("Quarta-Feira")
    case 5:
        print("Quinta-Feira")
    case 6:
        print("Sexta-Feira")
    case 7:
        print("Sábado-Feira")
    case _:
        print("Insira um número válido")
```

Menu de bebidas

```python
print("a) Café")
print("b) Chá")
print("c) Suco")
print("d) Água")

num = str(input("Digite uma opção:"))

match num:
    case "Café":
        print("Café")
    case "Chá":
        print("Chá")
    case "Suco":
        print("Suco")
    case "Água":
        print("Água")
    case _:
        print("Escolha uma opção válida")
```

Funções texto

```python
texto = str(input("Digite um texto:"))

texto_mod = texto.lower()
print(texto_mod)

texto_mod = texto.upper()
print(texto_mod)

texto_mod = texto.capitalize()
print(texto_mod)
```

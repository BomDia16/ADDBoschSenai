Calculadora

```python
num1 = float(input("Digite um número:"))
num2 = float(input("Digite um outro número:"))
escolha = int(input("Dgite qual opção deseja(1 - Soma, 2 - Multiplicação, 3 - Divisão)"))

if (escolha == 1):
    print("Soma =", num1 + num2)
elif (escolha == 2):
    print("Multiplicação =", num1 * num2)
elif (escolha == 3):
    if (num2 == 0):
        print("Não é possível fazer divisão por 0")
    else:
        print(f"Divisão = {(num1 / num2):,.2f}")
else:
    print("Insira um valor válido")
```

Ano de nascimento

```python
ano = int(input("Insira o ano de nascimento"))
ano_atual = 2025

idade = ano_atual - ano

if (idade < 16 and idade >= 0):
    print(f"Você ainda não pode votar, faltam {16 - idade} anos para poder votar")
elif ((idade >= 16 and idade < 18) or idade > 60):
    print("Você pode votar mas não é obrigatório")
elif (idade >= 18):
    print("É obrigatório votar")
else:
    print("Insira um ano válido")
    print(idade)
```

Notas para conceitos

```python
nota = int(input("Digite a nota do aluno:"))
conceito = ""

if (nota >= 90):
    conceito = "A (Excelente!)"
elif (nota >= 80):
    conceito = "B (Muito Bom)"
elif (nota >= 70):
    conceito = "C (Bom)"
elif (nota >= 60):
    conceito = "D (Regular)"
elif (nota < 60 and nota >= 0):
    conceito = "F (Reprovado)"
else:
    print("Insira um valor válido")

print(f"Conceito {conceito}")
```

Notas aluno

```python
nota = int(input("Digite a nota do aluno:"))
conceito = ""

if (nota < 3 and nota >= 0):
    conceito = "E"
elif (nota >= 3 and nota <= 5):
    conceito = "D"
elif (nota >= 6 and nota <= 7):
    conceito = "C"
elif (nota >= 8 and nota <= 9):
    conceito = "B"
elif (nota == 10):
    conceito = "A"
else:
    print("Insira um valor válido")

print(f"Nota é conceito {conceito}")
```

Notas para saber a média

```python
nota1 = float(input("Digite a primeira nota:"))
nota2 = float(input("Digite a segunda nota:"))

media = (nota1 + nota2) / 2

if (media >= 0 and media <= 4):
    print(f"Média: {media}, Reprovado")
elif (media > 4 and media <= 7):
    print(f"Média: {media}, Exame")
elif (media > 7 and media<= 10):
    print(f"Média: {media}, Aprovado")
else:
    print("Insira um valor válido")
```

Caminhão

```python
estado = int(input("Insira o código do estado (1-5)"))
toneladas = float(input("Peso da carga do caminhão em toneladas"))
carga = int(input("Insita o código da carga (10-40)"))
imposto = 0
preco_carga = 0

if ((estado > 0 and estado <= 5) and toneladas > 0 and (carga >= 10 and carga <= 40)):
    kilogramas = toneladas * 1000
    
    print(f"Peso da carga em Kg: {kilogramas}")
    
    if (carga >= 10 and carga <= 20):
        preco_carga = 200 * kilogramas
    elif (carga > 20 and carga <= 30):
        preco_carga = 350 * kilogramas
    elif (carga > 30 and carga <= 40):
        preco_carga = 440 * kilogramas
        
    print(f"Preço da carga: R$ {preco_carga}")
        
    if (estado == 1):
        imposto = 0.35
    elif (estado == 2):
        imposto = 0.25
    elif (estado == 3):
        imposto = 0.15
    elif (estado == 4):
        imposto = 0.05
    
    imposto_aplicado = preco_carga * imposto
        
    print(f"Imposto aplicado: R$ {imposto_aplicado}")
    
    print(f"Valor total da carga com imposto: R$ {imposto_aplicado + preco_carga}")

else:
    print("Insira valores válidos")
```

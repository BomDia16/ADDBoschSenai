```python
#importar biblioteca
import pandas as pd
#variável arquivo csv

arquivo = 'dados.csv'

var = pd.read_csv(arquivo, sep = ';')

print(var)
```

```python
# 5 primeiras head()

print(var.head())
```

```python
#5 ultimos tail()

print(var.tail())
```

```python
#resumo - quantidade de linhas e colunas shape

print(var.shape)
```

```python
#lista completa

print(var.to_string())
```

```python
#trazendo informações info()

print(var.info())
```

```python
#dropna()

print(var.dropna())
```

```python
#shape

print(var.shape)
```

```python
#describe - resumo estatístico

print(var.describe())
```

```python
#mostrar um elemento específico -> loc[]

print(var.loc[10])
```

```python
#mostrar intervalo  loc[]

print(var.loc[1:20])
```

```python
# mostrar linha separadas loc[[]]

print(var.loc[[1, 3]])
```

```python
#filtrar dados objeto.loc [objeto[]] 

print(var.loc[var['VALOREMPENHO'] < 200000])
```

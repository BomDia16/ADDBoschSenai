Cenário: Cada produto fabricado pela Bosch possui diversas características que precisam ser
armazenadas e consultadas eficientemente.
Instruções:
1. Crie um dicionário chamado detalhes_produto_bosch com as seguintes chaves e valores:
o 'id_produto': "BOSCH-XYZ-123"
o 'nome': "Sensor de Temperatura Inteligente"
o 'versao': 2.1
o 'fabricado_em': "Alemanha"
o 'componentes': ['chip_principal', 'resistor_termico', 'conector_usb'] (uma lista)
2. Imprima o dicionário completo.
3. Use keys() para obter uma visualização das chaves e imprima-as.
4. Use values() para obter uma visualização dos valores e imprima-os.
5. Use items() para obter uma visualização dos pares chave-valor e imprima cada par.

```python
detalhes_produto_bosch = {
    'id_produto': "BOSCH-XYZ-123",
    'nome': "Sensor de Temperatura Inteligente",
    'versao': 2.1,
    'fabricado_em': "Alemanha",
    'componentes': ['chip_principal', 'resistor_termico', 'conector_usb']
}

print(f"Detalhes completos do produto: {detalhes_produto_bosch}")
print(f"Chaves: {detalhes_produto_bosch.keys()}")
print(f"Valores: {detalhes_produto_bosch.values()}")
print(f"Itens (chave-valor): {detalhes_produto_bosch.items()}")
```

# Tarefa 2: Registro de Produção por Linha (com Built-ins)
Cenário: O controle de produção da fábrica registra a quantidade de peças produzidas por cada
linha de montagem diariamente.
Instruções:
1. Crie um dicionário chamado producao_por_linha com as seguintes chaves (nomes das
linhas) e valores (quantidade produzida):
o 'Linha A': 1200
o 'Linha B': 1500
o 'Linha C': 950
o 'Linha D': 1350
2. Imprima o dicionário.
3. Adicione uma nova linha 'Linha E' com 1100 peças. Verifique se ela foi adicionada.
4. Use o método pop() para remover a 'Linha C' e imprima a quantidade removida.
5. Imprima o dicionário atualizado.
6. Use len() para obter o número de linhas de produção ativas.
7. Use clear() para remover todos os itens do dicionário e imprima-o vazio.

```python
producao_por_linha = {
    'Linha A': 1200,
    'Linha B': 1500,
    'Linha C': 950,
    'Linha D': 1350
}

print(f"Produção inicial por linha: {producao_por_linha}")
producao_por_linha['Linha E'] = 1100
print(f"Produção após adicionar linha E: {producao_por_linha}")

print(f"\nRemovido Linha C: {producao_por_linha['Linha C']} peças produzidas.")
producao_por_linha.pop('Linha C')
print(f"Produção após remover Linha C: {producao_por_linha}")

print(f"\nNúmero total de linhas ativas: {len(producao_por_linha)}")

producao_por_linha.clear()
print(f"Produção após limpar: {producao_por_linha}")
```

# Tarefa 3: Gerenciamento de Carrinho de Compras em E-commerce
Cenário: Uma pequena loja virtual armazena os itens do carrinho de compras de um cliente usando
um dicionário aninhado. Cada item no carrinho possui detalhes como quantidade e preço unitário.
Instruções:
1. Crie um dicionário chamado carrinho_cliente.
2. Dentro deste dicionário, crie uma estrutura aninhada para representar os itens. Use o nome
do produto como chave principal e, para cada produto, tenha um dicionário com as chaves
quantidade e preco_unitario.
o Adicione ao carrinho: "Mouse Sem Fio" com quantidade: 2 e preco_unitario: 25.50.
o Adicione ao carrinho: "Teclado Mecânico" com quantidade: 3 e preco_unitario:
120.00.

3. Imprima o dicionário completo do carrinho.
4. Acesse e imprima a quantidade de "Mouse Sem Fio".
5. Calcule e imprima o preço total do "Teclado Mecânico”

```python
carrinho_cliente = {
    'Mouse Sem Fio': {
        'quantidade': 2,
        'preco_unitario': 25.5,
    },
    'Teclado Mecânico': {
        'quantidade': 2,
        'preco_unitario': 120.0
    }
}

print(f"Carrinho do Cliente:\n{carrinho_cliente}")

print(f"\nQuantidade de Mouse Sem Fio: {carrinho_cliente['Mouse Sem Fio']['quantidade']}")
preco_total_teclado = carrinho_cliente['Teclado Mecânico']['preco_unitario'] * carrinho_cliente['Teclado Mecânico']['quantidade']
print(f"Preço total do Teclado Mecânico: R$ {preco_total_teclado}")
```

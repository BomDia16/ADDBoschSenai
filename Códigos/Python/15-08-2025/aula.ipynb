## Tenha um excelente dia

Numero maior, menor ou igual a dez

```python
num = int(input("Insira um número inteiro: "))
if(num > 10):
    print("O número", num, "é maior que dez")
else:
    print("O número", num, "é menor ou igual dez")
```

Qual numero é maior entre dois

```python
num1 = int(input("Insira um número inteiro: "))
num2 = int(input("Insira um outro número inteiro: "))

if(num1 > num2):
    print("O número", num1, "é maior que", num2)
elif(num1 < num2):
    print("O número", num2, "é maior que", num1)
else:
    print("Os dois números são iguais")
```

Média de dois números

```python
num1 = float(input("Insira um número inteiro: "))
num2 = float(input("Insira um outro número inteiro: "))

media = (num1 + num2)/2

print(f"Primeira nota: {num1:,.2f}")
print(f"Segunda nota: {num2:,.2f}")
print(f"A média das notas é {media:,.2f}")
```

Divisão com dísima periódica

```python
num1 = float(input("Insira um número inteiro: "))
num2 = float(input("Insira um outro número inteiro: "))

dividir = num1 / num2

print(f"O resultado é {dividir:,.2f}")
```

1. Papelaria

```python
copias = int(input("Insira a quantidade de cópias:"))
total = 0

if (copias <= 100):
    total = 0.25 * copias
else:
    total = 0.2 * copias
    
print("Total a pagar: R$", total)
```

2. Número maior

```python
num1 = int(input("Insira um número inteiro: "))
num2 = int(input("Insira um outro número inteiro: "))

if(num1 > num2):
    print("O número", num1, "é maior que", num2)
elif(num1 < num2):
    print("O número", num2, "é maior que", num1)
else:
    print("Os dois números são iguais")
```

3. Maior que 20 ou não

```python
num = int(input("Insira um número inteiro: "))

if (num > 20):
    print(num / 2)
else:
    print(num * 2)
```

4. A+B é menor que C

```python
A = int(input("Insira um número inteiro: "))
B = int(input("Insira um outro número inteiro: "))
C = int(input("Insira mais um número inteiro: "))

if ((A + B) < C):
    print("A soma de A+B é menor que C")
else:
    print("A soma de A+B é maior ou iqual a C")
```

5. A e B maior ou menor

```python
A = int(input("Insira um número inteiro: "))
B = int(input("Insira um outro número inteiro: "))

if (A == B):
    print(A + B)
else:
    print(A * B)
```

6. Loja com parcelas

```python
preco = float(input("Insira o valor do produto:"))
parcelas = int(input("Insira o número de parcelas:"))

if (parcelas == 3):
    preco = preco * 1.1
elif (parcelas == 5):
    preco = preco * 1.2

preco_parcela = preco / parcelas

print(f"Preço total: R${preco:,.2f}")
print(f"Preço parcela: R${preco_parcela:,.2f}")
print("Quantidade de parcelas:", parcelas)
```

7. Notas aluno

```python
nome = str(input("Nome do aluno:"))
nota1 = float(input("Insira a nota 1:"))
nota2 = float(input("Insira a nota 2:"))
nota3 = float(input("Insira a nota 3:"))

media = (nota1 + nota2 + nota3) / 3

print("Aluno:", nome)
print(f"Média: {media:,.2f}")

if (media >= 7):
    print("APROVADO")
else:
    print("REPROVADO")
```

8. Subtração de valores

```python
num1 = int(input("Insira um número inteiro: "))
num2 = int(input("Insira um outro número inteiro: "))

if (num1 > num2):
    diferenca = num1 - num2
    print("A diferença é de", diferenca)
    
    if (diferenca > 50):
        print("A diferença entre os números é maior que 50")
    else:
        print("A diferença entre os números é menor ou igual a 50")
else:
    print("O primeiro número tem que ser maior que o segundo")
```

9. Alcool ou Gasolina

```python
alcool = float(input("Preço do alcool:"))
gasolina = float(input("Preço da gasolina:"))

calculo = alcool / gasolina

if (calculo < 0.7):
    print("O alcool vale mais a pena")
elif (calculo >= 0.7):
    print("A gasolina vale mais a pena")
```

10. Manutenção maquina

```python
produzidas = int(input("Total de peças:"))
defeituosas = int(input("Peças defeituosas:"))

if ((defeituosas / produzidas) > 0.1):
    print("A máquina precisa de manutenção")
else:
    print("A máquina está OK")
```

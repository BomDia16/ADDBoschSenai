7. Crie um programa que leia vários números positivos digitados pelo usuário, até que ele digite um valor negativo. Ao fim, o programa deve mostrar a média dos números positivos digitados.

```python
# use while
qtd = 0
soma = 0
while True:
    n = int(input("Insira um número positivo para adicionar a soma, insira um número negativo para parar:"))
    
    if(n < 0):
        if (qtd == 0):
            qtd = 1
        break

    soma += n
    qtd += 1
    
media = soma / qtd
print(media)
```

8. Faça uma tabuada de 1 a 10. O usuário digita o valor e a tabuada mostra os resultados. Crie uma flag para encerrar a tabuada.

```python
FINAL = 10
i = 1
num = int(input("Insira um número para ver a tabuada:"))

while i <= FINAL:
    print(i, "*", num, "=", i * num)
    i += 1
```

9. Crie um programa que calcule e mostre o fatorial de um número digitado pelo usuário.

```python
# use while
# Fatorial ->3! = 3 * 2 * 1 = 6

contador = int(input("Insira um valor para ver o seu fatorial:"))
fatorial = 1
while contador > 0:
    fatorial = contador * fatorial
    contador -= 1
print(fatorial)
```

```python
# use for
fatorial = 1
num = int(input("Insira um valor para ver o seu fatorial:"))
for i in range(num, 0, -1):
    fatorial = fatorial * i
print(fatorial)
```

11. Crie um programa que calcule e mostre todos os números primos de 1 a 100.

```python
for num in range(1, 101):
    if num < 2:
        continue
    primo = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            primo = False
            break
    if primo:
        print(num)
```

12. Sua tarefa é ajudar o mágico a preencher o código no editor de forma que o código:
* solicitará que o usuário insira um número inteiro;
* vai usar laço;
* verificará se o número inserido pelo usuário é igual ao número escolhido pelo mágico. Se o número escolhido pelo usuário for diferente do número secreto do mago, o usuário deverá ver a mensagem"Ha ha! Você está preso no meu loop!"Será solicitado a inserir um número novamente. Se o número inserido pelo usuário corresponde ao número escolhido pelo mago, ele deverá ser impresso na tela, e o mago deve dizer as seguintes palavras:"Muito bem! Você está livre agora."

```python
import random
secreto = random.randint(1, 10)
num = 0

while num != secreto:
    num = int(input("Insira um número inteiro para escapar do mágico:"))
    
    if num == secreto:
        print("Muito bem! Você está livre agora.")
    else:
        print("Ha ha! Você está preso no meu loop!")
```

13. Um 1937, um matemático alemão chamado Lothar Collatz formulou uma hipótese intrigante (ainda não comprovada) que pode ser descrita da seguinte forma:

* 2. se por par, avalie um novo c0 como c0 ÷ 2;

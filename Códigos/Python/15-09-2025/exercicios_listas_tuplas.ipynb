# Tarefa 1: Armazenamento e Exibição de Produção Diária
Cenário: O departamento de produção da Bosch regista o número de peças produzidas em cada
turno do dia.
Instruções:
1. Crie uma lista chamada producao_diaria com valores numéricos que representem a
produção de 3 turnos (ex: [500, 650, 480]).
2. Imprima a lista inteira.
3. Acesse e imprima a produção do segundo turno (índice 1).
4. Adicione um valor para o quarto turno (ex: 550) usando o método append().
5. Imprima a lista atualizada.

```python
producao_diaria = [500, 650, 480]
print(f'Produção diária: {producao_diaria}')
print(f"Produção do segundo turno: {producao_diaria[1]}")
producao_diaria.append(550)
print(f"Produção diária atualizada com o quarto turno: {producao_diaria}")
```

# Tarefa 2: Gerenciamento de Estoque de Matéria-Prima
Cenário: O sistema de estoque da Bosch precisa registar os tipos de matéria-prima disponíveis.
Instruções:
1. Crie uma lista chamada materiais_disponiveis com alguns strings (ex: ['aço', 'plástico',
'alumínio', 'borracha']).
2. Remova a borracha da lista usando o método remove().
3. Verifique se o plástico ainda está na lista usando o operador in e imprima uma mensagem
correspondente.
4. Crie uma nova lista com os materiais ordenados alfabeticamente.

```python
materiais_disponiveis = ['aço', 'plástico', 'alumínio', 'borracha']
print(f"Materiais antes da remoção: {materiais_disponiveis}")
materiais_disponiveis.remove("borracha")
print(f"Materiais após a remoção: {materiais_disponiveis}")
if 'plástico' in materiais_disponiveis:
    print("Plástico está disponível no estoque.")
materiais_disponiveis.sort()
lista_ordenada = materiais_disponiveis
print(f"Materiais ordenados: {lista_ordenada}")
```

# Tarefa 3: Análise de Dados de Sensores (Listas e Loop for)
Cenário: Uma lista de leituras de temperatura de um sensor foi coletada. O objetivo é calcular a
temperatura média e encontrar o valor máximo.
Instruções:
1. Crie uma lista de floats chamada leituras_temperatura (ex: [85.5, 87.2, 86.8, 89.0, 85.9]).
2. Use um loop for e a para soma de todas as leituras.
3. Calcule a média dividindo a soma pelo número de itens.
4. Encontre o valor máximo na lista usando a função max() do Python.
5. Imprima a média e o valor máximo formatando as saídas.

```python
leituras_temperatura = [85.5, 87.2, 86.8, 89.0, 85.9]
soma = 0

for i in leituras_temperatura:
    soma += i

media = soma / len(leituras_temperatura)
maximo = max(leituras_temperatura)

print(f"Temperatura média: {media}°C")
print(f"Temperatura máxima registrada: {maximo}°C")
```

# Tarefa 5: Coordenadas de Sensores com Tupla
Cenário: A posição de sensores instalados em uma área específica da fábrica pode ser representada
por coordenadas (X, Y). Um conjunto de sensores terá suas coordenadas armazenadas em tuplas.
Instruções:
1. Crie uma tupla chamada coordenadas_sensor_1 com os valores (50.5, 25.2).
2. Crie outra tupla chamada coordenadas_sensor_2 com os valores (75.0, 30.8).
3. Armazene essas duas tuplas dentro de uma lista. Assim, você terá uma lista de tuplas.
4. Imprima a lista completa de coordenadas.
5. Acesse a tupla de coordenadas do segundo sensor (índice 1 da lista) e imprima apenas a
coordenada X desse sensor (índice 0 da tupla).

```python
coordenadas_sensor_1 = (50.5, 25.2)
coordenadas_sensor_2 = (75.0, 30.8)

coordenadas = []
coordenadas.append(coordenadas_sensor_1)
coordenadas.append(coordenadas_sensor_2)
print(f"Lista de coordenadas dos sensores: {coordenadas}")
print(f"Coordenada X do segundo sensor: {coordenadas[1][0]}")
```

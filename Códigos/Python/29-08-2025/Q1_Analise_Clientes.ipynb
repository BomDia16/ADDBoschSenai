# Questão 1: Análise de Clientes e Ações de Vendas

Você recebeu uma lista de IDs de clientes, mas alguns estão duplicados devido a erros de registro e
outros precisam ser removidos após um cancelamento.

1. Crie uma lista chamada clientes_registrados contendo os seguintes IDs: ['C101', 'C102',
'C103', 'C104', 'C102', 'C105', 'C106', 'C101', 'C107'].

```python
clientes_registrados = ['C101', 'C102', 'C103', 'C104', 
                        'C102', 'C105', 'C106', 'C101', 'C107']
```

2. Use o método count() para verificar quantas vezes o ID 'C101' aparece na lista. Imprima o
resultado.

```python
contar = clientes_registrados.count('C101')
print(contar)
```

3. Um novo cliente, com ID 'C108', fez um registro. Use append() para adicioná-lo ao final da
lista.

```python
clientes_registrados.append('C108')
print(clientes_registrados)
```

4. O cliente com ID 'C103' cancelou o serviço. Use remove() para retirá-lo da lista.

```python
clientes_registrados.remove('C103')
print(clientes_registrados)
```

5. Use sort() para ordenar a lista em ordem alfabética e imprima a lista final.

```python
clientes_registrados.sort()
print(clientes_registrados)
```

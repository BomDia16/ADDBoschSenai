# Questão 4: Gerenciamento de Estoque de Componentes Eletrônicos
Uma lista de componentes eletrônicos foi registrada, mas contém itens duplicados e precisa ser
organizada para um inventário.

1. Crie uma lista inventario = ['Resistor', 'Capacitor', 'LED', 'Resistor', 'Transistor', 'Diodo',
'Capacitor', 'Resistor'].

```python
inventario = ['Resistor', 'Capacitor', 'LED', 'Resistor', 
              'Transistor', 'Diodo', 'Capacitor', 'Resistor']
```

2. Use count() para determinar quantas vezes o item 'Resistor' aparece e imprima o resultado.

```python
contar = inventario.count('Resistor')
print(contar)
```

3. Remova todas as ocorrências duplicadas de 'Resistor' da lista. (Dica: utilize um loop for ou
while e o método remove()).

```python
inventario_sem_duplicatas = []
vistos = set()

for item in inventario:
    if item not in vistos:
        inventario_sem_duplicatas.append(item)
        vistos.add(item)
        
inventario = inventario_sem_duplicatas

print(inventario)
```

4. Use append() para adicionar um novo componente, 'Microchip', ao inventário.

```python
inventario.append('Microchip')
print(inventario)
```

5. Use pop() para remover o último componente da lista.

```python
inventario.pop()
print(inventario)
```

6. Use sort() para ordenar a lista restante em ordem alfabética. Imprima a lista final.

```python
inventario.sort()
print(inventario)
```

# Conversor Celsius para Fahrenheit

```python
def conversor(celsius):
    fahrenheit = (celsius * 1.8) + 32
    return fahrenheit

print("A temperatura é de:", conversor(25), "°F")
```

# Calcular IMC

```python
def calcular(peso, altura):
    imc = peso / (altura * altura)
    print(classificar(imc))

def classificar(imc):
    if imc >= 40:
        return "Obseidade Grau III (Mórbida)"
    elif imc >= 35:
        return "Obesidade Grau II"
    elif imc >= 30:
        return "Obesidade Grau I"
    elif imc >= 25:
        return "Sobrepeso"
    elif imc >= 18.5:
        return "Peso normal"
    else:
        return "Magreza"
    
calcular(18.2, 1)
```

# Ano bissexto

```python
def ano_bissexto(ano):
    divisao_por_4 = lambda ano: ano % 4
    divisao_por_100 = lambda ano: ano % 100
    divisao_por_400 = lambda ano: ano % 400

    if divisao_por_4(ano) == 0:
        if divisao_por_400(ano) == 0:
            print(f"O ano {ano} é bissexto")
            return True
        elif divisao_por_100(ano) == 0:
            print(f"O ano {ano} não é bissexto")
            return False
        else:
            print(f"O ano {ano} é bissexto")
            return True
    else:
        print(f"O ano {ano} não é bissexto")
        return False

try:
    ano = int(input("Digite um ano"))
except ValueError:
    print("Isso não é um ano válido")
else:
    print(ano_bissexto(ano))
```

```python
somar = lambda x, y: x + y
subtrair = lambda x, y: x - y
multiplicar = lambda x, y: x * y
dividir = lambda x, y: x / y

print(somar(1, 3))
print(subtrair(1, 3))
print(multiplicar(1, 3))
print(dividir(1, 3))
```

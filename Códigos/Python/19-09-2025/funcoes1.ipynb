# Função de Calculadora

```python
def calculadora(num1, num2, operacao):
    resultado = ""
    match operacao:
        case '+':
            resultado = num1 + num2
        case '-':
            resultado = num1 - num2
        case '*':
            resultado = num1 * num2
        case '/':
            if num2 == 0:
                print("Não é possível realizar divisão por 0")
            else:
                resultado = num1 / num2
        case _:
            print("Operação inválida")
    
    if resultado != "":
        print("O resultado é igual a:", resultado)
    
calculadora(1, 1, "+")
calculadora(1, 1, "-")
calculadora(2, 2, "*")
calculadora(2, 0, "/")
calculadora(2, 2, "/")
calculadora(1, 1, "a")
```
